```python
import pandas as pd
import matplotlib.pyplot as plt

# Define the file path (update this based on your actual file location)
file_path = "API_SP.POP.TOTL_DS2_en_csv_v2_87.csv"

# Load the CSV file, skipping the first 4 rows (metadata)
df = pd.read_csv(file_path, skiprows=4)

# Extract 2023 population data and drop NaN values
df_population_2023 = df[['Country Name', '2023']].dropna()

# Convert population values to numeric
df_population_2023['2023'] = pd.to_numeric(df_population_2023['2023'], errors='coerce')

# Plot histogram of population distribution
plt.figure(figsize=(10, 6))
plt.hist(df_population_2023['2023'], bins=20, color='skyblue', edgecolor='black', alpha=0.7)
plt.xlabel('Population (log scale)')
plt.ylabel('Number of Countries')
plt.title('Histogram of Population Distribution (2023)')
plt.xscale('log')  # Using log scale for better visualization
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Show the plot
plt.show()
```
